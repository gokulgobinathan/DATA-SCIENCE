```python
import pandas as pd
import io

df = pd.read_csv("/content/PokemonData.csv")
df
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
ab=df['Type2'].mode()
ab
```

```python
df['Type2'].replace(np.nan,ab[0],inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df['Legendary'].replace(True,"Legendary",inplace=True)
df['Legendary'].replace(False,"Non-Legendary",inplace=True)
df
```

#graphs


```python
a=df['Legendary'].value_counts()
a.plot.barh(color=['blue','yellow'])
```

```python
import matplotlib.pyplot as plt

bins=range(0,200,20)
plt.hist(df["Attack"],bins,histtype="bar",rwidth=1.2,color='green')
plt.xlabel('Attack')
plt.ylabel('Count')
plt.plot()
plt.axvline(df['Attack'].mean(),linestyle='dashed',color='red')
plt.show()
```

```python
fire=df[(df['Type1']=='Fire') | ((df['Type2'])=="Fire")]
water=df[(df['Type1']=='Water') | ((df['Type2'])=="Water")]
plt.scatter(fire.Attack,fire.Defense,color='red',label='Fire')
plt.scatter(water.Attack,water.Defense,color='blue',label="Water")
plt.xlabel("Attack")
plt.ylabel("DEFENCE")
plt.legend()
plt.plot()
fig=plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

```python
import seaborn as sns
df2=df.drop(['Generation','HP'],axis=1)
sns.boxplot(data=df2)
plt.ylim(0,600)
plt.show()
```

```python
plt.subplots(figsize = (13,5))
plt.title('Attack by Type1')
sns.boxplot(x = "Type1", y = "Attack",data = df)
plt.ylim(0,200)
plt.show()
```
